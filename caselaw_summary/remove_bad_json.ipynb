```python
import os
import json
from collections import defaultdict

losses = defaultdict(int)

for model in ["gpt-3.5-turbo-1106", "gemma-7b-it", "mixtral-8x7b-32768"]:
    path = f"answers/{model}/"
    for prompt_type in os.listdir(path):
        for selection_type in os.listdir(path + prompt_type):
            for file in os.listdir(path + prompt_type + "/" + selection_type):
                print(model, prompt_type, selection_type, file)
                try:
                    j = json.load(open(path + prompt_type + "/" + selection_type + "/" + file, "r"))
                    if "usage" not in j[-1].keys():
                        os.remove(path + prompt_type + "/" + selection_type + "/" + file)
                        losses[model] += 1
                except Exception as e:
                    print(type(e).__name__)
                    os.remove(path + prompt_type + "/" + selection_type + "/" + file)
                    losses[model] += 1

losses

```
